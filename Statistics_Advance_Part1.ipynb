#Statistics Advance Part 1
1.What is a random variable in probability theory?
- A random variable is a numerical outcome of a random process or experiment. It assigns a real number to each possible outcome in a sample space, allowing probabilities to be analyzed mathematically. Random variables can be discrete or continuous, depending on the type of outcomes they represent.

2.What are the types of random variables?
- Random variables are of two main types: discrete and continuous. Discrete random variables take specific, separate values (like whole numbers), often from counting events. Continuous random variables can take any value within a range, often from measuring (like height or weight). Each type helps in modeling different kinds of data.

3.What is the difference between discrete and continuous distributions?
- A discrete distribution deals with countable outcomes (like rolling a die), while a continuous distribution involves outcomes over a range (like measuring height). Discrete values are separate and distinct; continuous values can take any value within an interval.

4.What are probability distribution functions (PDF)?
- A probability distribution function (PDF) describes how likely a random variable is to take on a specific value. For continuous variables, the PDF gives the probability density, and the area under the curve over an interval shows the actual probability for that range.

5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A probability distribution function (PDF) shows the likelihood of a random variable taking specific values, while a cumulative distribution function (CDF) gives the probability that the variable is less than or equal to a certain value. The CDF is the total area under the PDF curve up to that point.

6.What is a discrete uniform distribution?
- A discrete uniform distribution is a probability distribution where each possible outcome has an equal chance of occurring. It applies to a finite number of distinct values, like rolling a fair die where each number from 1 to 6 has the same probability of 1/6.

7.What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution models a random experiment with only two outcomes: success (1) and failure (0). It has one parameter, p, the probability of success. The mean is p, variance is p(1−p), and it's memoryless and discrete. It’s often used for binary events.

8.What is the binomial distribution, and how is it used in probability?
- The binomial distribution is a probability model used to find the chance of getting a fixed number of successes in a set number of trials, where each trial has only two outcomes (like success or failure) and the same chance of success. It’s useful in yes/no type experiments.

9.What is the Poisson distribution and where is it applied?
- The Poisson distribution is a probability model that describes how often a rare event occurs in a fixed time or space. It's used in fields like traffic flow, call centers, and biology—anywhere we count how many times something happens, like cars passing by or emails arriving.

10.What is a continuous uniform distribution?
- A continuous uniform distribution describes a situation where all outcomes within a certain range are equally likely. For example, if you randomly pick a number between 0 and 1, every number in this interval has the same probability of being chosen.

11.What are the characteristics of a normal distribution?
- A normal distribution is symmetrical, with most data points clustering around the mean. It follows a bell-shaped curve, where the mean, median, and mode are all equal. It has a characteristic spread determined by the standard deviation, and about 68% of values lie within one standard deviation from the mean.

12.What is the standard normal distribution, and why is it important?
- The standard normal distribution is a bell-shaped curve with a mean of 0 and a standard deviation of 1. It’s essential in statistics because it serves as a reference for measuring probabilities, making it easier to standardize and compare different data sets.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that, regardless of a population's distribution, the sampling distribution of the sample mean will be approximately normal if the sample size is large enough. It is crucial because it enables inferential statistics, allowing us to make predictions about populations based on sample data.

14.How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population's distribution. This connection highlights that many real-world phenomena can be modeled using the normal distribution, especially with large samples.

15.What is the application of Z statistics in hypothesis testing?
- Z statistics are used in hypothesis testing to determine if a sample mean significantly differs from a population mean, assuming the population is normally distributed. It's especially useful when the sample size is large (n > 30) or the population standard deviation is known. It helps assess the probability of obtaining a result under the null hypothesis.

16.How do you calculate a Z-score, and what does it represent?
- To calculate a Z-score, subtract the mean from the data point and divide by the standard deviation: Z = (X - μ) / σ. It represents how many standard deviations a value is from the mean of the dataset. A positive Z-score means above the mean, while a negative one means below.

17.What are point estimates and interval estimates in statistics?
- A point estimate is a single value used to estimate an unknown population parameter, like the sample mean as an estimate of the population mean. An interval estimate, on the other hand, provides a range of values (e.g., confidence intervals) within which the parameter is likely to lie.

18.What is the significance of confidence intervals in statistical analysis?
- Confidence intervals show the range in which we expect the true value to lie, with a certain level of confidence (like 95%). They help estimate uncertainty in results and allow better decision-making by showing how precise or reliable a statistical measure is.

19.What is the relationship between a Z-score and a confidence interval?
- A Z-score tells how far a data point is from the mean in standard deviations. Confidence intervals use Z-scores to define the range around a sample mean where the true population mean likely lies. Higher confidence levels require larger Z-scores, making the interval wider.

20.How are Z-scores used to compare different distributions?
- Z-scores standardize values by showing how far they are from the mean in standard deviations. This allows comparison between different distributions, even if they have different units or scales, by placing them on the same relative scale.

21.What are the assumptions for applying the Central Limit Theorem?
- The Central Limit Theorem (CLT) assumes that samples are independent, identically distributed, and drawn from a population with a finite mean and variance. For non-normal populations, a large enough sample size (usually n ≥ 30) is needed for the sample mean to approximate a normal distribution.

22.What is the concept of expected value in a probability distribution?
- Expected value is the average outcome you’d expect from a probability distribution if an experiment were repeated many times. It’s found by multiplying each possible outcome by its probability and then adding them all together. It gives a single number summarizing the distribution’s center.

23.How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution shows how likely each value of a random variable is. The expected outcome, or expected value, is the average you’d get if you repeated the random process many times. It’s calculated using the distribution by multiplying each value by its probability and adding them up.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculates the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The value (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: The probability of observing x
    """
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

```python
#8.Simulate data from a normal distribution and plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0  # Mean (mu)
std_dev = 1  # Standard deviation (sigma)
size = 1000  # Number of samples to generate

# Generate samples from normal distribution
samples = np.random.normal(mean, std_dev, size)

# Plot the histogram of the distribution
plt.hist(samples, bins=30, density=True, edgecolor='black', alpha=0.7)

# Plot the theoretical PDF of the normal distribution
x = np.linspace(min(samples), max(samples), 1000)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, pdf, 'r', lw=2, label="Theoretical PDF")

# Title and labels
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
    data (array-like): The dataset (list or numpy array).

    Returns:
    np.array: The Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 1000)  # Generate 1000 samples from a normal distribution

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.plot(z_scores, label='Z-scores')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution(n, p, size):
    """
    Generate random variables using a binomial distribution and calculate their probabilities.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success on each trial.
    size (int): Number of random variables to generate.

    Returns:
    np.array: Array of generated random variables.
    np.array: Array of corresponding probabilities.
    """
    # Generate random variables from the binomial distribution
    random_vars = np.random.binomial(n, p, size)

    # Calculate the probabilities for each outcome (from 0 to n)
    outcomes = np.arange(0, n + 1)
    probabilities = binom.pmf(outcomes, n, p)

    return random_vars, outcomes, probabilities

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success on each trial
size = 1000  # Number of random variables to generate

# Generate random variables and calculate probabilities
random_vars, outcomes, probabilities = binomial_distribution(n, p, size)

# Plot the histogram of the generated random variables
plt.hist(random_vars, bins=np.arange(n+2)-0.5, edgecolor='black', density=True, alpha=0.7, label="Generated Random Variables")

# Plot the PMF of the binomial distribution
plt.plot(outcomes, probabilities, 'ro', label="Theoretical PMF", lw=2)

# Title and labels
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(0, n+1))
plt.legend()
plt.grid(True)
plt.show()
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a given sample data.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The mean of the population (under the null hypothesis).
    population_std (float): The standard deviation of the population.
    alpha (float): Significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, p-value, decision to reject or not reject the null hypothesis.
    """
    # Sample statistics
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed p-value

    # Decision rule
    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

# Sample data (for example, test scores from a sample of students)
sample_data = np.array([85, 87, 90, 92, 88, 95, 91, 84, 89, 93])

# Population parameters (e.g., population mean and standard deviation)
population_mean = 90  # Hypothesized population mean
population_std = 5    # Known population standard deviation

# Perform Z-test
z_stat, p_value, reject_null = z_test(sample_data, population_mean, population_std)

# Display results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
if reject_null:
    print("Reject the null hypothesis: There is enough evidence to support the alternative hypothesis.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to support the alternative hypothesis.")

# Plotting the Z
```

```python
#20.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
    data (array-like): The sample data.
    confidence_level (float): The confidence level for the interval (default is 0.95 for 95%).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use sample standard deviation (ddof=1)
    n = len(data)

    # Z-score for the desired confidence level (for a normal distribution)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Standard error
    standard_error = sample_std / np.sqrt(n)

    # Margin of error
    margin_of_error = z_score * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example data: a sample of test scores
sample_data = np.array([85, 87, 90, 92, 88, 95, 91, 84, 89, 93])

# Calculate the 95% confidence interval
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

# Display the result
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#21.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

# Function to calculate the confidence interval
def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
    data (array-like): The sample data.
    confidence_level (float): The confidence level for the interval (default is 0.95 for 95%).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use sample standard deviation (ddof=1)
    n = len(data)

    # Z-score for the desired confidence level (for a normal distribution)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Standard error
    standard_error = sample_std / np.sqrt(n)

    # Margin of error
    margin_of_error = z_score * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50   # Mean of the normal distribution
std_dev = 10  # Standard deviation of the normal distribution
size = 100  # Sample size

# Generate random data
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate the 95% confidence interval for the mean
lower, upper = confidence_interval(data, confidence_level=0.95)

# Display the result
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev, x_range=(-10, 10), num_points=1000):
    """
    Calculate and visualize the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.
    x_range (tuple): The range of x values (default is (-10, 10)).
    num_points (int): The number of points to sample in the range for the plot (default is 1000).
    """
    # Generate x values within the specified range
    x = np.linspace(x_range[0], x_range[1], num_points)

    # Calculate the PDF of the normal distribution
    pdf = norm.pdf(x, mean, std_dev)

    # Plot the PDF
    plt.plot(x, pdf, label=f"Normal Distribution\n(μ={mean}, σ={std_dev})", color='blue')
    plt.title(f"Normal Distribution PDF (μ={mean}, σ={std_dev})")
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example: Mean = 0, Standard Deviation = 1
mean = 0
std_dev = 1

# Call the function to plot the PDF
plot_normal_pdf(mean, std_dev)
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda, k_max):
    """
    Plot the CDF of a Poisson distribution for a given lambda and range of k.

    Parameters:
    lmbda (float): The average rate (lambda) of the Poisson distribution.
    k_max (int): The maximum value of k for which we will calculate the CDF.
    """
    # Calculate the CDF for each k from 0 to k_max
    k_values = np.arange(0, k_max + 1)
    cdf_values = poisson.cdf(k_values, lmbda)

    # Plot the CDF
    plt.step(k_values, cdf_values, where='post', label=f'Poisson CDF (λ={lmbda})', color='blue')
    plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example parameters for Poisson distribution
lmbda = 3  # Average rate (lambda)
k_max = 10  # Maximum number of events to calculate CDF for

# Plot the CDF of the Poisson distribution
plot_poisson_cdf(lmbda, k_max)

# Calculate CDF for specific k values
k_values_to_check = [0, 1, 2, 3, 4]
cdf_results = {k: poisson.cdf(k, lmbda) for k in k_values_to_check}

# Display CDF results for the selected k values
print(f"CDF Results for Poisson Distribution with λ = {lmbda}:")
for k, cdf_val in cdf_results.items():
    print(f"P(X ≤ {k}) = {cdf_val:.4f}")
```

```python
#24. Simulate a random variable using a continuous uniform distribution and calculate its expected value.
import numpy as np

def simulate_uniform_distribution(a, b, size=1000):
    """
    Simulate random variables from a continuous uniform distribution and calculate the expected value.

    Parameters:
    a (float): The lower bound of the distribution.
    b (float): The upper bound of the distribution.
    size (int): Number of random variables to simulate (default is 1000).

    Returns:
    tuple: The simulated random variables and the calculated expected value.
    """
    # Simulate random variables from the uniform distribution
    random_variables = np.random.uniform(a, b, size)

    # Calculate the expected value of the uniform distribution (mean of the bounds)
    expected_value = (a + b) / 2

    return random_variables, expected_value

# Define the range for the uniform distribution
a = 0  # Lower bound
b = 10  # Upper bound

# Simulate the random variables and calculate the expected value
random_vars, expected_value = simulate_uniform_distribution(a, b)

# Display the expected value and the first 10 simulated random variables
print(f"Expected Value (Theoretical): {expected_value}")
print(f"First 10 Simulated Random Variables: {random_vars[:10]}")

# Optionally, you can also calculate the sample mean of the simulated data
sample_mean = np.mean(random_vars)
print(f"Sample Mean (Simulated): {sample_mean}")
```

```python
#25.Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1 (array-like): First dataset.
    data2 (array-like): Second dataset.
    """
    # Calculate the standard deviations of the two datasets
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation
    std_dev2 = np.std(data2, ddof=1)  # Sample standard deviation

    # Print the standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Visualize the difference using a bar chart
    labels = ['Dataset 1', 'Dataset 2']
    std_devs = [std_dev1, std_dev2]

    # Create a bar chart
    plt.bar(labels, std_devs, color=['blue', 'green'])
    plt.title('Comparison of Standard Deviations')
    plt.xlabel('Datasets')
    plt.ylabel('Standard Deviation')
    plt.ylim(0, max(std_devs) * 1.2)  # Add some padding above the bars
    plt.show()

# Example datasets
np.random.seed(42)  # For reproducibility
data1 = np.random.normal(loc=50, scale=10, size=100)  # Normally distributed data
data2 = np.random.normal(loc=50, scale=20, size=100)  # Normally distributed data with higher variance

# Compare the standard deviations of the two datasets
compare_standard_deviations(data1, data2)
```

```python
#26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a given dataset.

    Parameters:
    data (array-like): The dataset to calculate the range and IQR for.

    Returns:
    tuple: The range and the interquartile range (IQR).
    """
    # Calculate the range
    data_range = np.max(data) - np.min(data)

    # Calculate the interquartile range (IQR)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    iqr = Q3 - Q1

    return data_range, iqr

# Generate a sample dataset from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50
std_dev = 10
size = 1000

# Generate random data
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate the range and IQR for the generated data
data_range, iqr = calculate_range_and_iqr(data)

# Display the results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Normalize the dataset using Z-score normalization.

    Parameters:
    data (array-like): The dataset to normalize.

    Returns:
    np.array: The Z-score normalized data.
    """
    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Generate a sample dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)

# Apply Z-score normalization
normalized_data = z_score_normalization(data)

# Visualize the original and normalized data
plt.figure(figsize=(12, 6))

# Original Data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized Data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a given dataset.

    Parameters:
    data (array-like): The dataset to calculate skewness and kurtosis for.

    Returns:
    tuple: Skewness and kurtosis of the dataset.
    """
    # Calculate skewness
    skewness = skew(data)

    # Calculate kurtosis (Fisher's definition, subtracting 3 for normal distribution)
    kurt = kurtosis(data, fisher=True)

    return skewness, kurt

# Generate a sample dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate skewness and kurtosis for the generated data
skewness, kurt = calculate_skewness_and_kurtosis(data)

# Display the results
print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
